Gathering crime rate for each state in the introduction year, previous year and after year. The crimes considered are rape_legacy, rape_revised, robbery, property_crime, burglary, larceny, motor_vehicle_theft. The reason for considering only these crimes as they are petty. - PIVOT POINT

```python
def crime_stats(state,year):
    crime_current = us_crime_df.loc[(us_crime_df['year'] == year) & (us_crime_df['state_name'] == state)]
    crime_current = crime_current.drop(columns=['state_abbr', 'population', 'violent_crime', 'homicide','aggravated_assault', 'caveats'])
    crime_before = us_crime_df.loc[(us_crime_df['year'] == year-1) & (us_crime_df['state_name'] == state)]
    crime_before = crime_before.drop(columns=['year', 'state_abbr', 'population', 'violent_crime', 'homicide','aggravated_assault', 'caveats'])
    crime_before.columns = ['state_name','before_rape_legacy', 'before_rape_revised', 'before_robbery', 'before_property_crime', 'before_burglary', 'before_larceny', 'before_motor_vehicle_theft']
    crime_after = us_crime_df.loc[(us_crime_df['year'] == year+1) & (us_crime_df['state_name'] == state)]
    crime_after = crime_after.drop(columns=['year', 'state_abbr', 'population', 'violent_crime', 'homicide','aggravated_assault', 'caveats'])
    crime_after.columns = ['state_name','after_rape_legacy', 'after_rape_revised', 'after_robbery', 'after_property_crime', 'after_burglary', 'after_larceny', 'after_motor_vehicle_theft']
    merged_df = pd.merge(crime_current, crime_before, on="state_name")
    df = pd.merge(merged_df, crime_after, on="state_name")
    return df
```

```python
main_df = pd.DataFrame()
for index, row in uber_introduction_date_df.iterrows():
    return_df = crime_stats(row['State'], row['Intro_year'])
    main_df = main_df.append(return_df)
print(main_df)
```

Calculating the number of states that had reduction in crime rate before UBER/LYFT introduction and after the introduction. Final reduction represents the count of the states that had a reduction in crime rate only after UBER/LYFT introduction and not before.

```python
column_names = ["Crime", "Before_Reduced_true", "Before_Reduced_false", "After_Reduced_true", "After_Reduced_false", "Final_Reduced_true", "Final_Reduced_false"]
percent_df = pd.DataFrame(columns = column_names)
for column in ['rape_legacy','rape_revised','robbery','property_crime','burglary','larceny','motor_vehicle_theft']:
    before_name = 'before_'+column+'_percent'
    after_name = 'after_'+column+'_percent'
    before_reduction = 'before_'+column+'_reduction'
    after_reduction = 'after_'+column+'_reduction'
    main_df[before_reduction] = np.where((main_df[before_name] < 0), 1, 0)
    main_df[after_reduction] = np.where((main_df[after_name] < 0), 1, 0)
    x = main_df[before_reduction].value_counts()
    y = main_df[after_reduction].value_counts()
    final_reduction = 'final_'+column+'_reduction'
    main_df[final_reduction] = np.where(main_df[after_reduction] == 1, np.where(main_df[before_reduction] == 0,1, 0), 0)
    z = main_df[final_reduction].value_counts()
    new_row = {'Crime':column, 'Before_Reduced_false':x[0], 'Before_Reduced_true':x[1], 'After_Reduced_false':y[0], 'After_Reduced_true': y[1], 'Final_Reduced_false': z[0], 'Final_Reduced_true': z[1]}
    percent_df = percent_df.append(new_row, ignore_index=True)
percent_df
```
